```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
```

```python
df = df.fillna('')
df
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('_./')
df
```

```python
df = df.drop( columns = 'Not_Useful_Column')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].replace('No','N')
df['Paying Customer'] = df['Paying Customer'].replace('N/a','')
df
```

```python
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
df["Phone_Number"] = df["Phone_Number"].str.replace('Na','', regex = False)
df["Phone_Number"] = df["Phone_Number"].str.replace('nan','', regex = False)
df["Phone_Number"] = df['Phone_Number'].apply(lambda x : str(x))
df["Phone_Number"] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace('--','', regex = False)
df
# df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
```

```python
df['Address'] = df['Address'].str.replace('N/a','')
df[['Street','State','Pincode']] = df['Address'].str.split(',',expand = True)
df
```

```python
df = df.fillna('')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('No','N')
# df['Paying Customer'] = df['Paying Customer'].replace('N/a','')
df
```

```python
df = df[df['Do_Not_Contact'] != 'Y']
df
```
